# Preprocessing Data (training and testing datasets)

## Preprocessing for Training Data

```python
import cv2
import os
from PIL import Image

# training dataset map for file handling
training_dataset_paths = {
    "glioma_tumor": "Dataset/Training/glioma_tumor",
    "meningioma_tumor": "Dataset/Training/meningioma_tumor",
    "no_tumor": "Dataset/Training/no_tumor",
    "pituitary_tumor": "Dataset/Training/pituitary_tumor"
}

# new width and height
new_width = 256
new_height = 256
```

```python
# training dataset

glioma_tumor_training_path = "Dataset/Training/glioma_tumor"
meningioma_tumor_training_path = "Dataset/Training/meningioma_tumor"
no_tumor_training_path = "Dataset/Training/no_tumor"
pituitary_tumor_training_path = "Dataset/Training/pituitary_tumor"

# get all files in the glioma_tumor directory
file_list = os.listdir(glioma_tumor_training_path)
# iterate through the files and rename them
for idx, file_name in enumerate(file_list, 1):
    file_path = os.path.join(glioma_tumor_training_path, file_name)
    
    # check if the path is a file
    if os.path.isfile(file_path):
        # define the new file name
        new_file_name = f"glioma_tumor_training({idx}){os.path.splitext(file_name)[1]}"
        new_file_path = os.path.join(glioma_tumor_training_path, new_file_name)
        
        # rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {file_name} to {new_file_name}")


# get all files in the meningioma_tumor directory
file_list = os.listdir(meningioma_tumor_training_path)
# iterate through the files and rename them
for idx, file_name in enumerate(file_list, 1):
    file_path = os.path.join(meningioma_tumor_training_path, file_name)
    
    # check if the path is a file
    if os.path.isfile(file_path):
        # define the new file name
        new_file_name = f"meningioma_tumor_training({idx}){os.path.splitext(file_name)[1]}"
        new_file_path = os.path.join(meningioma_tumor_training_path, new_file_name)
        
        # rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {file_name} to {new_file_name}")


# get all files in the no_tumor directory
file_list = os.listdir(no_tumor_training_path)
# iterate through the files and rename them
for idx, file_name in enumerate(file_list, 1):
    file_path = os.path.join(no_tumor_training_path, file_name)
    
    # check if the path is a file
    if os.path.isfile(file_path):
        # define the new file name
        new_file_name = f"no_tumor_training({idx}){os.path.splitext(file_name)[1]}"
        new_file_path = os.path.join(no_tumor_training_path, new_file_name)
        
        # rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {file_name} to {new_file_name}")


# get all files in the pituitary_tumor directory
file_list = os.listdir(pituitary_tumor_training_path)
# iterate through the files and rename them
for idx, file_name in enumerate(file_list, 1):
    file_path = os.path.join(pituitary_tumor_training_path, file_name)
    
    # check if the path is a file
    if os.path.isfile(file_path):
        # define the new file name
        new_file_name = f"pituitary_tumor_training({idx}){os.path.splitext(file_name)[1]}"
        new_file_path = os.path.join(pituitary_tumor_training_path, new_file_name)
        
        # rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {file_name} to {new_file_name}")
```

```python
# resize all the training dataset images to (256, 256) pixels
for tumor_type, path in training_dataset_paths.items():
    for file_name in os.listdir(path):
        # get file path
        file_path = os.path.join(path, file_name)
        # image from path
        image = cv2.imread(file_path)
        if image is not None:
            resized_image = cv2.resize(image, (new_width, new_height))
            cv2.imwrite(file_path, resized_image)
            print(f"Resized: {file_path}")
        else:
            print(f"Could not read image: {file_path}")
```

```python
# normalizing the image pixels to set pixel in range [0, 1]
for tumor_type, path in training_dataset_paths.items():
    for file_name in os.listdir(path):
        file_path = os.path.join(path, file_name)
        image = cv2.imread(file_path)
    
        # check if image is properly read
        if image is not None:
            # normalize the image pixels to [0, 1]
            normalized_image = image / 255.0

            # save the normalized image back to the file path (in uint8 format)
            normalized_image = (normalized_image * 255).astype('uint8')
            cv2.imwrite(file_path, normalized_image)
            print(f"Normalized: {file_path}")
        else:
            print(f"Could not read image: {file_path}")
```

# Extract Features

```python
# flattening images for training dataset images

import numpy as np

flattened_images_train = {}

# iterate through image files
for tumor_type, path in training_dataset_paths.items():
    for file_name in os.listdir(path):
        file_path = os.path.join(path, file_name)
        image = cv2.imread(file_path)
        if image is not None:
            # flatten image to 1 dimensional image
            flattened_image = image.flatten()
            
            # add the flattened image to the hashmap
            if tumor_type not in flattened_images_train:
                flattened_images_train[tumor_type] = [flattened_image]
            else:
                flattened_images_train[tumor_type].append(flattened_image)

# testing to make sure count is correct

# for list in flattened_images.values():
    # print(len(list))
```

```python
# flattening images for testing dataset images

import numpy as np

flattened_images_test = {}

# iterate through image files
for tumor_type, path in testing_dataset_paths.items():
    for file_name in os.listdir(path):
        file_path = os.path.join(path, file_name)
        image = cv2.imread(file_path)
        if image is not None:
            # flatten image to 1 dimensional image
            flattened_image = image.flatten()
            
            # add the flattened image to the hashmap
            if tumor_type not in flattened_images_test:
                flattened_images_test[tumor_type] = [flattened_image]
            else:
                flattened_images_test[tumor_type].append(flattened_image)

# testing to make sure count is correct

# for list in flattened_images.values():
    # print(len(list))
```

# Prepare the Data for Training

```python
# convert image labels to numerical format

label_map = {
    "glioma_tumor": 0,
    "meningioma_tumor": 1,
    "no_tumor": 2,
    "pituitary_tumor": 3
}
```

# Model Selection, Training, and Evaluation

## Convert Data to NumPy Array (training dataset)

```python
# training dataset

import numpy as np

# convert training data to NumPy arrays
X_train = []
Y_train = []

for tumor_type, images in flattened_images_train.items():
    for image in images:
        X_train.append(image)
        Y_train.append(label_map[tumor_type])

# convert to NumPy arrays
X_train = np.array(X_train)
Y_train = np.array(Y_train)

print(X_train)
print(Y_train)
```

## Convert Data to NumPy Array (testing dataset)

```python
# testing dataset

# convert testing data to NumPy arrays
X_test = []
Y_test = []

for tumor_type, images in flattened_images_test.items():
    for image in images:
        X_test.append(image)
        Y_test.append(label_map[tumor_type])

X_test = np.array(X_test)
Y_test = np.array(Y_test)

print(X_test)
print(Y_test)
```

## Train K-NN Model and Evaluate

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define class names
classes = ["glioma", "meningioma", "no_tumor", "pituitary"]

# Define k values
k_values = [3, 5, 7, 9, 11]

# Create a figure with separate rows for each k value
fig, axes = plt.subplots(len(k_values), 4, figsize=(12, len(k_values) * 4))  # 1 row per k, 4 columns (1 per class)

for row_idx, k in enumerate(k_values):
    # Train KNN model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)

    # Create a binary confusion matrix for each class
    for col_idx, class_name in enumerate(classes):
        binary_Y_test = (Y_test == col_idx).astype(int)  # Convert to binary (1 = current class, 0 = others)
        binary_Y_pred = (Y_pred == col_idx).astype(int)  # Convert to binary

        # Compute confusion matrix
        cm = confusion_matrix(binary_Y_test, binary_Y_pred)

        # Handle cases where only one class is present
        if cm.shape == (1, 1):  
            cm = np.array([[cm[0][0], 0], [0, 0]])

        # Ensure matrix is 2x2 (TP, FN, FP, TN)
        if cm.shape == (2, 2):
            TP, FN = cm[0]
            FP, TN = cm[1]
        else:
            TP = FN = FP = TN = 0  # Edge case handling

        # Print the confusion matrix explicitly
        print(f"\nEvaluation Scores for {class_name} (k={k}):")
        # print(f"TP: {TP}, FN: {FN}")
        # print(f"FP: {FP}, TN: {TN}")
        print(f"Accuracy:  {(TP + TN) / (TP + TN + FP + FN)}")
        print(f"Precision: {TP / (TP + FP)}")
        print(f"Recall:    {TP / (TP + FN)}")
        print(f"F-measure: {(2 * TN) / ((2 * TN) + FN + FP)}")

        # Plot confusion matrix
        ax = axes[row_idx, col_idx]  # Select the subplot
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=[class_name, "Other"], yticklabels=[class_name, "Other"], ax=ax)
        ax.set_xlabel("Predicted Label")
        ax.set_ylabel("True Label")
        ax.set_title(f"{class_name} (k={k})")

# Adjust layout
plt.tight_layout()
plt.show()
```

## Train Decision Tree Model and Evaluate
